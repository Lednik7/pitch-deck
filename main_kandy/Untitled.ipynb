```python
from PIL import Image
import urllib.request

prompt = """Slidesgo, Freepik. Pinterest. More white. Maybe gradient. WITHOUT TEXT. Vectors style. WITHOUT TEXT. Image for PowerPoint background. Gradientdip. More game with colors. smooth transition. smooth transition. smooth transition"""
URL = f"http://141.105.64.158:8730/images"

headers = {
    "query": prompt,
    "negative": ""
}

req = urllib.request.Request(URL, headers=headers)
with urllib.request.urlopen(req) as url:
    img = Image.open(url)
    img.show()
```
